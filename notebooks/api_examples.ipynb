## Data + features

```python
from car_pricing import data, features
df = data.clean(data.load_raw())
X, y = features.split_xy(df)
edges = features.band_edges(y)
print('shape:', X.shape, '| edges:', edges)
features.price_to_band([2.0, 5.0, 25.0], edges)
```
